```python
import pandas as pd
import numpy as np
```

### Series
Series是一个包含一系列值的一维数组对象和一个关联的数据标签（index）  
pd.Series类似于np.array，pd.Series主要是用于构建含有标签的一维数组

```python
obj = pd.Series([4, 7, -5, 3], index=list('ABCD'), name='score', dtype=np.int16)
obj
```

Series.values 获取Series的值  
Series.index 获取Series的索引


```python
obj.values, obj.index
```

相比于Numpy，可以使用标签对单个值或一组值进行访问

```python
obj['A'], obj[['A', 'C', 'D']]
```

使用Boolean数组进行过滤

```python
obj2 = pd.Series([4, 7, -5, 3], index=['d', 'b', 'a', 'c'])
```

```python
np.exp(obj2)
```

Series可以看做是字典，所以可以用in来判断是否包含有个标签，以及可以通过字典来创建Series

```python
sdata = {'Ohio': 35000, 'Texas': 71000, 'Oregon': 16000, 'Utah': 5000}
obj3 = pd.Series(sdata)
obj3
```

isnull和notnull用于检查是否为缺失值

```python
data = {'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada', 'Nevada'],
'year': [2000, 2001, 2002, 2001, 2002, 2003],
'pop': [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]}

frame = pd.DataFrame(data, columns=['year', 'state', 'pop'], index=['one', 'two', 'three', 'four', 'five', 'six'])
frame
```

```python
val = pd.Series([-1.2, -1.5, -1.7], index=['two', 'three', 'four'])
frame['debt'] = val
frame
```

```python
frame = pd.DataFrame(np.arange(9).reshape((3, 3)),
                      index=['a', 'c', 'd'],
                      columns=['Ohio', 'Texas', 'California'])

frame2 = frame.reindex(['a', 'b', 'c', 'd'], method='ffill')
frame2
```

```python
frame2['Ohio']
```
